```python
import numpy as np
import time
```

```python
a = np.array([1,2,3,4,5])
b = 2 * a
b
```

```python
def dot_product(a,b):
    x = 0
    n = a.shape[0]
    for i in range(n):
        x+=a[i] + b[i]
    return x
    
```

```python
tic = time.time()
dot_product(a,b)
toc = time.time()
print('Duration of the calculation using for loop is', toc-tic)

```

```python
tic = time.time()
np.dot(a,b)
toc = time.time()
print('Duration of the calculation using dot product is', toc-tic)
```
