TRAINING POLYNOMIAL REGRESSION MODEL

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import gradient_descent, cost_function , compute_derivative
```

Training datas

```python
x=np.arange(0,20,1)
y = 1 + x**2
```

```python
x= x.reshape(-1,1)  #here we are making each datas in x as the training examples
m,n = x.shape
```

Initial parameters of the model

```python
w_init = np.zeros(n)   #the number of weight depends upon the number of features
b_init = 0
```

Training our model 

```python
alpha = 1e-2
iterations = 1000
w,b,loss_error,hist_iterations = gradient_descent(x,y,w_init,b_init,compute_derivative,cost_function,alpha,iterations)
x.shape
```

As we can observe that the relation between the training example and the output is not linear , as its ploynomial, more like a curve,
so our linear regression model doesnot fit properly in this relation and is not good fit for the prediction of output

```python
plt.scatter(x,y,label='original data values',c='b')
plt.plot(x,(x@w) + b,label='Predicted data values',c='r')
plt.legend()
plt.show()
```

Feature engineering,
As we can observe from the plot above that the relation is more like a curve, so we must do some feature engineering,
So that our regression model will fit efficiently , inorder to predict the data

```python
x_second=np.arange(0,20,1)
y_second = 1 + x_second**2
x_second = x_second**2  #engineering the feature
x_second = x_second.reshape(-1,1)
m,n = x_second.shape
```

Training the polyregression model

```python
w_init = np.zeros(n)   #the number of weight depends upon the number of features
b_init = 0
alpha = 1e-5
iterations = 10000
w_second,b_second,loss_error,hist_iterations = gradient_descent(x_second,y_second,w_init,b_init,compute_derivative,cost_function,alpha,iterations)
w_second
```

Plotting the polynomial regression model

```python
plt.scatter(x_second,y_second,label='original data values',c='b')
plt.plot(x_second,np.dot(x_second,w_second) + b_second,label='Predicted data values',c='r')
plt.legend()
plt.show()
```

Polynomial feature expansion

```python
x_third = np.arange(0,20,1)
y_third = x_third ** 2
x_third = np.c_[x_third,x_third**2,x_third**3]  #feature engineering
```

```python
m,n = x_third.shape
w_init = np.zeros(n)   #the number of weight depends upon the number of features
b_init = 0
alpha = 1e-7
iterations = 10000
w_third,b_third,loss_error,hist_iterations = gradient_descent(x_third,y_third,w_init,b_init,compute_derivative,cost_function,alpha,iterations)
print(f'The weight of the model is {w_third} and the bias of the model is: {b_third}')
```
