TRAINING POLYNOMIAL REGRESSION MODEL

```python
import numpy as np
import matplotlib.pyplot as plt
from utils import gradient_descent, cost_function , compute_derivative
```

Training datas

```python
x=np.arange(0,20,1)
y = 1 + x**2
```

```python
x= x.reshape(-1,1)  #here we are making each datas in x as the training examples
m,n = x.shape
```

Initial parameters of the model

```python
w_init = np.zeros(n)   #the number of weight depends upon the number of features
b_init = 0
```

Training our model 

```python
alpha = 1e-2
iterations = 1000
w,b,loss_error,hist_iterations = gradient_descent(x,y,w_init,b_init,compute_derivative,cost_function,alpha,iterations)
x.shape
```

As we can observe that the relation between the training example and the output is not linear , as its ploynomial, more like a curve,
so our linear regression model doesnot fit properly in this relation and is not good fit for the prediction of output

```python
plt.scatter(x,y,label='original data values',c='b')
plt.plot(x,(x@w) + b,label='Predicted data values',c='r')
plt.legend()
plt.show()
```
