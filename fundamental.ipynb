```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# x_train is the input variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
#this is the ith training example
i = 0
x_i = x_train[i]
y_i = y_train[i]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
```

```python
#length of training examples
m = x_train.shape[0]
m  #this is the length of the training example
```

```python
plt.figure(figsize=(5,2))
plt.scatter(x_train,y_train,marker='x',c='r')
plt.xlabel('House sqaure ft.')
plt.ylabel('House Price')
plt.title('House Price Prediction')
plt.show()
```

```python
#using random weight and bias
w = 200
b = 100 
```

```python
def compute_model_output(x,w,b):
    m = x.shape[0]  #number of training datas
    output = np.zeros(m)
    for i in range(m):
        output[i] = w * x[i] + b
    return output
tmp_f_wb = compute_model_output(x_train, w, b,)    
print(tmp_f_wb)        
        
```

```python
plt.plot(x_train,tmp_f_wb,label='Predicted values')
plt.scatter(x_train,y_train,marker='x',c='r',label='Original values')
plt.title('Housing price prediction')
plt.xlabel('House sqr ft.')
plt.ylabel('House Price')
plt.legend()
plt.show()
```
