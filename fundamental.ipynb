```python
#using random weight and bias
w = 200
b = 100 
```

```python
def compute_model_output(x,w,b):
    m = x.shape[0]  #number of training datas
    output = np.zeros(m)
    for i in range(m):
        output[i] = w * x[i] + b
    return output
tmp_f_wb = compute_model_output(x_train, w, b,)    
print(tmp_f_wb)        
        
```

```python
plt.plot(x_train,tmp_f_wb,label='Predicted values')
plt.scatter(x_train,y_train,marker='x',c='r',label='Original values')
plt.title('Housing price prediction')
plt.xlabel('House sqr ft.')
plt.ylabel('House Price')
plt.legend()
plt.show()
```

```python
def compute_derivative(w,x,y,m):  #here this function calculates the partial derivative
    dj_dw = 0  #partial derivative with respect to weight
    dj_db = 0  #partial derivative with respect to the bias
    for i in range(m):
        y_pred_i = w * x[i] + b
        dj_dw += (y_pred_i - y[i]) * x[i]
        dj_db += (y_pred_i - y[i])
    grad = {
        'djw':(1/m) * (dj_dw),
        'djb':(1/m) *  dj_db
    }    
    return grad
print(compute_derivative(w,x_train,y_train,m))  

        
    
```

```python
def cost_function(w,x,y,b):
    m=x.shape[0]
    squared_error = 0
    for i in range(m):
        y_pred = w * x[i] + b
        squared_error += np.square(y_pred - y[i])
    cost = (1/2*m) * squared_error
    return cost
        
```

```python
def gradient_descent(w,b,x,y,m,gradient_function,iterations,alpha):
    cost_values = []
    parameters_history = []
    for i in range(iterations):
        grad = gradient_function(w,x,y,m)
        djw = grad['djw']
        djb = grad['djb']
        w = w - alpha * djw
        b = b - alpha * djb
        cost = cost_function(w,x,y,b)
        if i<100000:
            cost_values.append(cost)
            parameters_history.append([w,b])
        print('The current cost function is:', cost)
    parameters = {
        'w':w,
        'b':b
    }    
    return parameters
w_init = 0
b_init = 0
# some gradient descent settings
iterations = 10000
tmp_alpha = 1.0e-2
print(gradient_descent(w_init,b_init,x_train,y_train,m,compute_derivative,iterations,alpha=1.0e-2))    
    
    
    
```
